```python
def translateString(string):
    tran = collections.OrderedDict()
    tran[u'а']='a'
    tran[u'б']='b'
    tran[u'в']='v'
    tran[u'г']='g'
    tran[u'д']='d'
    tran[u'е']='e'
    tran[u'ж']='zh'
    tran[u'з']='z'
    tran[u'и']='i'
    tran[u'й']='y'
    tran[u'к']='k'
    tran[u'л']='l'
    tran[u'м']='m'
    tran[u'н']='n'
    tran[u'о']='o'
    tran[u'п']='p'
    tran[u'р']='r'
    tran[u'с']='s'
    tran[u'т']='t'
    tran[u'у']='u'
    tran[u'ф']='f'
    tran[u'х']='kh'
    tran[u'ц']='ts'
    tran[u'ч']='ch'
    tran[u'ш']='sh'
    tran[u'щ']='shch'
    tran[u'ъ']=''
    tran[u'ы']='y'
    tran[u'ь']=''
    tran[u'э']='e'
    tran[u'ю']='yu'
    tran[u'я']='ya'
    tran[u'А']='A'
    tran[u'Б']='B'
    tran[u'В']='V'
    tran[u'Г']='G'
    tran[u'Д']='D'
    tran[u'Е']='E'
    tran[u'Ж']='Zh'
    tran[u'З']='Z'
    tran[u'И']='I'
    tran[u'Й']='Y'
    tran[u'К']='K'
    tran[u'Л']='L'
    tran[u'М']='M'
    tran[u'Н']='N'
    tran[u'О']='O'
    tran[u'П']='P'
    tran[u'Р']='R'
    tran[u'С']='S'
    tran[u'Т']='T'
    tran[u'У']='U'
    tran[u'Ф']='F'
    tran[u'Х']='Kh'
    tran[u'Ц']='Ts'
    tran[u'Ч']='Ch'
    tran[u'Ш']='Sh'
    tran[u'Щ']='Shch'
    tran[u'Ъ']=''
    tran[u'Ы']='Y'
    tran[u'Ь']=''
    tran[u'Э']='E'
    tran[u'Ю']='Yu'
    tran[u'Я']='Ya'
    tran[' ']=' '
    tran['-']='-'
    tran['.']='.'
    tran["'"]=''
    tran["’"]=''
    tran['(']='('
    tran[')']=')'
    tran['_']='_'
    charList = list(string)
    newString = ""
    for c in charList:  
        if c==u'ё':
            c=u'е'
        if c==u'Ё':
            c=u'Е'
        if c in tran:
            newString+=tran[c]
        else:
            if re.search('[a-zA-Z]',c):
                newString+=c
            else:
                print "translate exception: ", string
                return 'Exception'
    return newString
```

```python
def nameFormat(string1):

    d1 = string1.find("(")
    d2 = string1.find(")")
    if d1>=0 and d2>=0:
        string1 = string1[:d1]+string1[d2+1:]

    string = re.split('[\.|\\s]',string1) 
    string = filter(None, string)
    #print string

    patInd = -1
    nameInd = -1
    surnameInd = -1
    
    if len(string)==3: #Surname Name Pat
        for ind in range(len(string)):
            if len(string[ind])==1 or string[ind] in ["Zh", "Kh","Ts", "Ch", "Sh", "Shch", "Yu", "Ya"]: #N.
                patInd = ind # last one word
        if patInd!=-1: #we found one char word        
            nameInd = patInd-1 #previous is name
        else: #no short words
            for ind in range(len(string)):
                if len(string[ind])>=4: #vich, vna #we try to find pat
                    if "vich" in string[ind][-4:] or "vna" in string[ind][-4:]:
                        patInd = ind
                        if ind!=0: #voynovich! ))
                            break
        if patInd>0: #we found pat        
            nameInd = patInd-1  #name is previous    


        for ind in range(len(string)):
            if ind!=patInd and ind!=nameInd: #surname is the rest index
                surnameInd = ind 
                break

    else: # not 3 words
        if len(string)==2: #if 2 words
            for ind in range(len(string)): #Surname N.
                if len(string[ind])==1 or string[ind] in ["Zh", "Kh","Ts", "Ch", "Sh", "Shch", "Yu", "Ya"]: # we try to find name
                    nameInd = ind
                    break

            if nameInd==-1: #Surname Name
                for ind in range(len(string)):
                    for names in namesDB:
                        if string[ind]==names or SequenceMatcher(None, string[ind], names).ratio()>=0.8: #if name in database or similar to it
                            nameInd = ind
                            break
                        if nameInd>=0:
                            break


            for ind in range(len(string)):
                if ind!=nameInd:
                    surnameInd = ind #surname
                    break

    if surnameInd == -1:
        print "Format Exception: ", string1
        return "Exception"
    else:
        if patInd ==-1:
            return string[surnameInd]+' '+string[nameInd][0]+'.'
        else:
            return string[surnameInd]+' '+string[nameInd][0]+'. '+string[patInd][0]+'.'
        
```

```python
def distrPlot(k, w, title, name):
    k = filter(lambda a: a != 0, k) #del zero values
    plt.close('all')
    fig, ax = plt.subplots(nrows=1, figsize=(7,4))
    fig.suptitle('Degree distribution of the '+title, fontsize=12)  
    ax.set_ylabel('Num of authors')
    ax.set_xlabel('Degree')
    bins = np.arange(min(k),max(k),w)
    hist, bins = np.histogram(k, bins=bins)
    width = np.diff(bins)
    center = (bins[:-1] + bins[1:]) / 2
    ax.bar(center, hist, align='center', width=width)
    ax.set_xticks(bins)
    ax.set_xticklabels([b for b in bins],rotation=90, fontsize=8)
    plt.tight_layout()
    fig.subplots_adjust(top=0.93)
    #plt.show()
    fig.savefig("/Users/olgagerasimova/Desktop/Data/figs/Distribution_"+name+".png")
```

```python
def basicStat(k, ws, row):
    col = 2
    ws.cell(row = row, column = col, value = sum(k))
    col+=1
    ws.cell(row = row, column = col, value = np.mean(k))
    col+=1
    ws.cell(row = row, column = col, value = max(k))
    col+=1
    k = filter(lambda a: a != 0, k) #del zero values
    ws.cell(row = row, column = col, value = len(k))
    col+=1
    ws.cell(row = row, column = col, value = np.mean(k))
    col+=1
    return ws
```

```python
#save nodelist edgelist
def NodesAndEdgesList(G, name):
    wb = Workbook() 
    ws = wb.active
    row = 1
    col = 1

    ws.cell(row=row, column=col, value='ID')
    col+=1
    ws.cell(row=row, column=col, value='Label')

    row+=1

    for n in G.nodes():
        col = 1
        ws.cell(row=row, column=col, value=n)
        col+=1
        ws.cell(row=row, column=col, value=n)
        row+=1

    wb.save('Gephi\NL_'+name+'.xlsx')
    wb = Workbook() 
    ws = wb.active
    row = 1
    col = 1

    ws.cell(row=row, column=col, value='source')
    col+=1
    ws.cell(row=row, column=col, value='target')
    col+=1
    ws.cell(row=row, column=col, value='weight')
    row+=1

    for ed in G.edges():
        col = 1
        ws.cell(row=row, column=col, value=ed[0])
        col+=1
        ws.cell(row=row, column=col, value=ed[1])
        col+=1
        ws.cell(row=row, column=col, value=G.get_edge_data(ed[0], ed[1], default=0).values()[0])
        row+=1

    wb.save('Gephi\EL_'+name+'.xlsx')
```

```python
#similarity
CosineSim = dict()
for n1 in G.nodes():
    x=sorted(nx.all_neighbors(G,n1))
    d1=len(x)
    for n2 in G.nodes():
        y=sorted(nx.all_neighbors(G,n2))
        d2 = len(y)
        z=sorted(nx.common_neighbors(G,n1,n2))
        num1 = len(z)
        if d1*d2!=0:
            CosineSim[(n1,n2)]=num1/float(math.sqrt(d1*d2))
        else:
            CosineSim[(n1,n2)]=-1
```
