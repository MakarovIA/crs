### Data collection

#### Downloading database of books, journals, conferences and etc. from http://www.scimagojr.com

```python
year = [i for i in range(2013,2017)]
newb = xlwt.Workbook()
news = newb.add_sheet('1')
row = 0
#inds_base = []
for y in year:
    response = urllib.urlopen("http://www.scimagojr.com/journalrank.php?year="+str(y))
    text = response.read().decode("utf-8")
    num_records = re.findall(r'\d+',re.findall(r'total_size=\d+', text)[0])[0]    #,re.findall(r"\"journalrank.php\?page=2&total_size=\d+\"",text)[0]                 
    num_page = np.ceil(float(num_records)/50)
    print y, ":",num_records, num_page

    f = 1
    while(f < num_page + 1):
        response = urllib.urlopen("http://www.scimagojr.com/journalrank.php?year="+str(y)+"&page="+str(f)+"&total_size="+str(num_records))
        text = response.read().decode("utf-8")
        table = re.findall(r'<tbody>[\w\W\s]+</tbody>', text)
        elems = re.findall(r"<a href=\"journalsearch.php\?q=[\d]+&amp;tip=sid&amp;clean=0\" title=\"view journal details\">[`±~!_&\'\;\.\/,\w\s\(\)\=\?\\*\+\[\]\{\}\"\^\|\\%#№@§\$:-]+</a>", unidecode(table[0]).encode("ascii"))
        kol = 0
        for e in elems:
            l1 = re.findall(r"\"journalsearch.php\?q=[\d]+&amp;tip=sid&amp;clean=0\"", e)
            l2 = re.sub(r'amp;','', l1[0])
            ind = re.findall(r"\d+", l2)
            if int(ind[0]) not in inds_base:
                inds_base.append(int(ind[0]))
                title = re.findall(r'>[`±~!_&\'\;\.\/,\w\s\(\)\=\?\*\+\[\]\{\}\"\^\|\\%#№@§\$:-]+</a>', e)
                news.write(row,0, ind[0])
                news.write(row,1,title[0][1:-4])
                news.write(row,2,"http://www.scimagojr.com/"+l2[1:-1])
                row += 1
                kol +=1
        print f, "from", num_page, "(",kol,")"
        f +=1
```
