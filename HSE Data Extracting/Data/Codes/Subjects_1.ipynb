```python
from xlrd import *
from collections import *
Scopus_subject = "Scopus_subject.xls"
book = open_workbook(Scopus_subject)
```

```python
sheet = book.sheet_by_index(0)
scopus_subject_to_id_dict = {}
scopus_id_to_subject_dict = OrderedDict()
for i in range(2, sheet.nrows):
    scopus_id_to_subject_dict.update({ int(sheet.cell(i, 1).value) :sheet.cell(i, 0).value}) 
    scopus_subject_to_id_dict.update({ sheet.cell(i, 0).value :int(sheet.cell(i, 1).value) })
```

```python
temp=OrderedDict(scopus_id_to_subject_dict)
for i in scopus_id_to_subject_dict.keys():
    if (i % 100 == 0): 
        del temp[i]
        
scopus_id_to_subject_dict=temp
```

```python
cat_length=len(scopus_id_to_subject_dict.keys())
def scopus_cat_bin(cat_list, scopus_id_to_subject_dict, cat_length):
    bin=[0]*cat_length
    for i in xrange(len(cat_list)):
        bin[scopus_id_to_subject_dict.keys().index(cat_list[i])]=1
    return bin
```
